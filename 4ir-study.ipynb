```python

import pandas as pd
import matplotlib.pyplot as plt

# Create DataFrame
data = {
    "age": ["30-40", "30-40", "30-40", "30-40", "30-40", "30-40", "40-50", "40-50", "40-50", "40-50", "40-50", "50-60", "50-60"],
    "gender" : ["F", "F", "M", "M", "M", "M", "F", "F", "M", "M", "M", "M", "M"],
    "rating": [4, 3, 4, 3, 2, 1, 2, 1, 3, 2, 1, 2, 1],
    "count": [1, 1, 1, 2, 5, 5, 1, 1, 5, 4, 14, 5, 20]
}
df = pd.DataFrame(data)

# Create FacetGrid
g = sns.FacetGrid(df, col="age", row="gender", margin_titles=True)
g.map_dataframe(sns.scatterplot, x="rating", y="count")
g.set_axis_labels("Rating", "Count")
g.add_legend()

plt.show()
```
